````python
MODELS_TO_TEST = {
    "DeepSeek-Coder-1.3B": "deepseek-ai/deepseek-coder-1.3b-instruct",
    "Phi-2-2.7B": "microsoft/phi-2",
    "Gemma-2B-IT": "google/gemma-2b-it",
    "Stable-Code-3B": "stabilityai/stable-code-3b",
}

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

def clean_generated_code(text, model_path):
    model_path = model_path.lower()
    if "gemma" in model_path:
        text = re.sub(r"<start_of_turn>user\n.*?<end_of_turn>\n<start_of_turn>model\n", "", text, flags=re.DOTALL).replace("<end_of_turn>", "")
    elif "phi-2" in model_path:
        text = re.sub(r"Instruct:.*\nOutput:", "", text, flags=re.DOTALL)
    else:
        text = re.sub(r"### Instruction:\n.*?\n\n### Response:", "", text, flags=re.DOTALL)
    match = re.search(r"```python\n(.*?)\n```", text, re.DOTALL)
    if match:
        text = match.group(1)
    return text.strip()

def is_syntactically_valid(code_string):
    if not code_string: return False
    try:
        ast.parse(code_string)
        return True
    except SyntaxError:
        return False

def calculate_advanced_metrics(code_string):
    if not is_syntactically_valid(code_string):
        return {"complexity": None, "maintainability": None, "loc": None}
    try:
        complexity = sum([c.complexity for c in cc_visit(code_string)]) if cc_visit(code_string) else 0
        maintainability = mi_visit(code_string, multi=True)
        loc = analyze(code_string).loc
        return {"complexity": complexity, "maintainability": round(float(maintainability),2), "loc": loc}
    except:
        return {"complexity": None, "maintainability": None, "loc": None}

def generate_code(model, tokenizer, prompt):
    model_path = model.name_or_path.lower()

    if "gemma" in model_path:
        formatted_prompt = f"<start_of_turn>user\n{prompt}<end_of_turn>\n<start_of_turn>model\n"
    elif "phi-2" in model_path:
        formatted_prompt = f"Instruct: {prompt}\nOutput:"
    else:
        formatted_prompt = f"### Instruction:\n{prompt}\n\n### Response:"

    if tokenizer.pad_token is None:
        tokenizer.pad_token = tokenizer.eos_token

    inputs = tokenizer(formatted_prompt, return_tensors="pt").to(device)

    gen_config = GenerationConfig(
        max_new_tokens=512,
        temperature=0.1,
        do_sample=True,
        pad_token_id=tokenizer.pad_token_id
    )

    start_time = time.time()
    output_ids = model.generate(
        inputs.input_ids,
        attention_mask=inputs.attention_mask,
        generation_config=gen_config
    )
    end_time = time.time()

    raw_output = tokenizer.batch_decode(output_ids, skip_special_tokens=True)[0]
    cleaned_code = clean_generated_code(raw_output, model_path)

    return {"code": cleaned_code, "gen_time": end_time - start_time}
````

````python
def load_model(model_path):
    """Load model + tokenizer just in time."""
    print(f"🚀 Loading model: {model_path}")
    tokenizer = AutoTokenizer.from_pretrained(model_path, trust_remote_code=True)
    model = AutoModelForCausalLM.from_pretrained(
        model_path,
        torch_dtype=torch.bfloat16,
        device_map="auto",
        trust_remote_code=True
    )
    return model, tokenizer

def unload_model(model, tokenizer):
    """Unload and clear GPU memory."""
    try:
        del model
        del tokenizer
        gc.collect()
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
        print("🧹 Model unloaded and GPU cache cleared.\n")
    except Exception as e:
        print(f"⚠️ Error unloading model: {e}")
````

````python
all_results_log = []

prompt_input_all = widgets.Textarea(
    placeholder='Enter a prompt to benchmark all models...', layout={'width':'95%'}
)
run_all_button = widgets.Button(description='Run Benchmark', button_style='danger', icon='rocket')
output_all = widgets.Output(layout={'border':'1px solid black','padding':'10px','overflow':'scroll'})

def on_run_all_clicked(b):
    with output_all:
        output_all.clear_output(wait=True)
        prompt = prompt_input_all.value.strip()
        if not prompt:
            print("Enter a prompt!"); return

        print(f"Running on {len(MODELS_TO_TEST)} models...\n")
        results_this_run = []

        for name, path in MODELS_TO_TEST.items():
            print(f"🧠 Running {name}...")
            try:
                model, tokenizer = load_model(path)
                result = generate_code(model, tokenizer, prompt)
                metrics = calculate_advanced_metrics(result['code'])
                entry = {'Model': name, 'Prompt': prompt, **result, **metrics}
                results_this_run.append(entry)
                all_results_log.append(entry)
            except Exception as e:
                print(f"❌ Error with {name}: {e}")
                continue
            finally:
                unload_model(model, tokenizer)

        print("✅ All models completed!\n")
        display(HTML(pd.DataFrame(results_this_run).round(2).to_html().replace('\\n','<br>')))

run_all_button.on_click(on_run_all_clicked)
display(widgets.VBox([prompt_input_all, run_all_button, output_all]))
````

````python
prompt_input_selected = widgets.Textarea(
    placeholder='Prompt for selected models...', layout={'width':'95%'}
)
run_selected_button = widgets.Button(description='Run Selected', button_style='success', icon='play')
output_selected = widgets.Output(layout={'border':'1px solid black','padding':'10px','overflow':'scroll'})

model_checkboxes = {name: widgets.Checkbox(value=True, description=name) for name in MODELS_TO_TEST.keys()}
checkbox_container = widgets.VBox(list(model_checkboxes.values()))

# Define the list of prompts to test
prompts_to_test = [
    "Write a Python program to perform merge sort on a list.",
    "Write a Python program to read a CSV file and print the first 5 rows.",
    "Write a Flask API with one /predict endpoint that returns a dummy JSON.",
    "Write Python code to connect to MySQL and fetch data from a students table.",
    "Write Python code to delete all .tmp files in a directory."
]


def on_run_selected_clicked(b):
    with output_selected:
        output_selected.clear_output(wait=True)

        models_to_run = [name for name, cb in model_checkboxes.items() if cb.value]
        if not models_to_run:
            print("Select at least one model!"); return

        print(f"Running {len(models_to_run)} selected models on {len(prompts_to_test)} prompts...\n")
        results_this_run = []

        for prompt in prompts_to_test:
            print(f"🎯 Running benchmark for prompt: {prompt}\n")
            for name in models_to_run:
                path = MODELS_TO_TEST[name]
                print(f"⚙️ Running {name}...")
                try:
                    model, tokenizer = load_model(path)
                    result = generate_code(model, tokenizer, prompt)
                    metrics = calculate_advanced_metrics(result['code'])
                    entry = {'Model': name, 'Prompt': prompt, **result, **metrics}
                    results_this_run.append(entry)
                    all_results_log.append(entry)
                except Exception as e:
                    print(f"❌ Error with {name}: {e}")
                    continue
                finally:
                    unload_model(model, tokenizer)

        print("✅ Completed selected models on all prompts!\n")
        if results_this_run:
            display(HTML(pd.DataFrame(results_this_run).round(2).to_html().replace('\\n','<br>')))
        else:
            print("No results were generated.")


run_selected_button.on_click(on_run_selected_clicked)
display(widgets.VBox([
    widgets.HTML("<h4>Select models:</h4>"),
    checkbox_container,
    run_selected_button,
    output_selected
]))
````

````python
report_button = widgets.Button(description="Generate Report & Plots", button_style='info')
report_output = widgets.Output()

def on_report_button_clicked(b):
    with report_output:
        report_output.clear_output(wait=True)
        if not all_results_log:
            print("No results to report."); return
        df = pd.DataFrame(all_results_log).round(2)
        df.rename(columns={
            'complexity':'Complexity',
            'maintainability':'Maintainability',
            'gen_time':'Gen Time (s)'
        }, inplace=True)
        display(df)

        sns.set_theme(style="whitegrid")
        plot_df = df.dropna(subset=['Complexity','Maintainability'])
        if plot_df.empty:
            print("Not enough data for plots"); return

        fig, axes = plt.subplots(1,3, figsize=(20,6))
        fig.suptitle('Code Generation Metrics', fontsize=16)

        sns.barplot(ax=axes[0], data=plot_df, x='Model', y='Gen Time (s)', palette='viridis')
        axes[0].set_title('Generation Time'); axes[0].tick_params(axis='x', rotation=45)

        sns.barplot(ax=axes[1], data=plot_df, x='Model', y='Complexity', palette='magma')
        axes[1].set_title('Cyclomatic Complexity'); axes[1].tick_params(axis='x', rotation=45)

        sns.barplot(ax=axes[2], data=plot_df, x='Model', y='Maintainability', palette='coolwarm')
        axes[2].set_title('Maintainability Index'); axes[2].tick_params(axis='x', rotation=45)

        plt.show()

report_button.on_click(on_report_button_clicked)
display(report_button, report_output)
````
